# 과제: 전자신문에서 인공지능 검색

### _-필요한 라이브러리 import_

```python
import numpy as np                      
import pandas as pd                    
import requests

import warnings
warnings.filterwarnings('ignore')
```

### _- params option 작성_

```python
# 전자신문 홈페이지에서 '인공지능' 을 검색했을 경우
url = 'https://search.etnews.com/etnews/search.php?kwd=%EC%9D%B8%EA%B3%B5%EC%A7%80%EB%8A%A5'

res = requests.get(url)

res.text
```

```python
# 아래 url 문자열 중에서 "?" 이하가 parameter 에 해당함
url = 'https://search.etnews.com/etnews/search.php'

params = { 
    'kwd': '인공지능'
}

res = requests.get(url, params=params).text 

res
```

### - _데이터 수집 후 해당 사이트 인코딩 방식 확인_

```python
url = 'https://search.etnews.com/etnews/search.php?kwd=%EC%9D%B8%EA%B3%B5%EC%A7%80%EB%8A%A5'

res = requests.get(url)    # get 메소드와  text 속성을 동시 사용 

res.encoding
```

### _-Status_code를 이용하여 에러여부 확인_

```python
url = 'https://search.etnews.com/etnews/search.php?kwd=%EC%9D%B8%EA%B3%B5%EC%A7%80%EB%8A%A5'

res = requests.get(url)

print("응답 코드: ", res.status_code)  # 응답코드가 200 이면 정상, 에러가 발생했다면 400번대 코드
print('\n')

if res.status_code == requests.codes.ok :
    print("정상입니다")
else:
    print("에러가 발생했습니다. [에러코드 ", res.status_code, "]")
```

### _-수집된 데이터 사이즈 확인 및 데이터 수집 후 내용 확인_

```python
print(len(res.text))
print('\n')
print(res.text)
```
