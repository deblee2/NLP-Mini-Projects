### 트위터에서 품사처리를 함

```python
twitter = Twitter()  #트위터만

morphs = [] 

for sentence in lists: 
    morphs.append(twitter.pos(sentence)) #트위터의 품사를 뽑아낸다

print(morphs) #morphs 리스트는 list of list
#list = set (set[0] == 단어, set[1] == 품사)
```

### 명사, 형용사, 부사만

```python
noun_adj_adv_list=[] 

for sentence in morphs : 
    for word, tag in sentence : #set에서 word, tag만 구분
        if tag in ['Noun']:  #a명사인 것만 포함 
            noun_adj_adv_list.append(word)

print(noun_adj_adv_list)
```

```python
from collections import Counter #모아서 저장하는 것. 없으면 for문과 dictionary, list를 사용해야함

count = Counter(noun_adj_adv_list)

words = dict(count.most_common()) #dictionary가 된다

words #{key: value}
```

### 워드 클라우드로 처리

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 자신의 컴퓨터 환경에 맞는 한글 폰트 경로를 설정
font_path = 'C:/Windows/Fonts/malgun.ttf'  #한글은 무조건 폰트 지정해야함

wc = WordCloud(width = 1500, height = 1000, background_color="white", colormap = "Accent_r", font_path=font_path)

plt.imshow(wc.generate_from_frequencies(words)) #dictionary에 들어있는 단어들의 frequency
plt.axis("off")
plt.show()
```

### 꼬꼬마로 처리

1. 꼬꼬마는 try except가 없으면 오류가 발생
2. 자바 null point exception 오류
3. 외부 오류이니 오류 문제도 그냥 Exception으로 처리하고 pass함

```python
#섬세한 구분을 위해서는 꼬꼬마
kkma = Kkma()

morphs_ = [] 

for sentence in lists: 
    try:
        morphs_.append(kkma.pos(sentence))
    except Exception as ex:
        pass
    
print(morphs_)    
```

```python
#명사 위주의 구분
noun_adj_adv_list=[] 

for sentence in morphs_ : 
    for word, tag in sentence : 
        if tag in ['NNG'] or tag in ['NNP'] or tag in ['MAG'] or tag in ['VA']: 
            noun_adj_adv_list.append(word) 

print(noun_adj_adv_list)
```

```python
from collections import Counter

count = Counter(noun_adj_adv_list)

words = dict(count.most_common())

words
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 자신의 컴퓨터 환경에 맞는 한글 폰트 경로를 설정
font_path = 'C:/Windows/Fonts/malgun.ttf' 

wc = WordCloud(width = 1500, height = 1000, background_color="black", colormap = "Accent_r", font_path=font_path)

plt.imshow(wc.generate_from_frequencies(words))
plt.axis("off")
plt.show()
```

### 박근혜, 이명박, 노무현 취임사를 분석함

```python
from konlpy.tag import *

#파일 열기
file = open("박근혜.txt", 'r') 
lists = file.readlines() 
file.close() 

#트위터
twitter = Twitter()  #트위터만

morphs = [] 

for sentence in lists: 
    morphs.append(twitter.pos(sentence))

#명사 형용사 부사
noun_adj_adv_list=[] 

for sentence in morphs : 
    for word, tag in sentence : #set에서 word, tag만 구분
        if tag in ['Noun']:  #a명사인 것만 포함 
            noun_adj_adv_list.append(word)

#Dictionary 형태로 모아서 저장
from collections import Counter #모아서 저장하는 것. 없으면 for문과 dictionary, list를 사용해야함

count = Counter(noun_adj_adv_list)

words = dict(count.most_common()) #dictionary가 된다 #{key: value}

#워드클라우드
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 자신의 컴퓨터 환경에 맞는 한글 폰트 경로를 설정
font_path = 'C:/Windows/Fonts/malgun.ttf' 

wc = WordCloud(width = 1500, height = 1000, background_color="black", colormap = "Accent_r", font_path=font_path)

plt.imshow(wc.generate_from_frequencies(words))
plt.axis("off")
plt.show()
```
